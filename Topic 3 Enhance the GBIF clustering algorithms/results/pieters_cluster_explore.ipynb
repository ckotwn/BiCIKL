### A summary of Meise data

Let's explore what the Meise Botanic Garden herbarium specimens link to/with using the GBIF clustering algorithm/dataset

### What institutions does Meise's herbarium cluster with

```
%sql
CREATE table pieter.external_occurrence
STORED AS parquet
AS SELECT r.reasons, o.*
  FROM gbif.relationships_triple r
    JOIN gbif.occurrence o ON r.id2 = o.gbifID
WHERE r.dataset1='b740eaa0-0679-41dc-acb7-990d562dfa37'    
```

```
%sql
select publisher, count(*) c
FROM pieter.external_occurrence 
GROUP BY publisher
ORDER BY c DESC
```

There is a long tail

```
%sql
select publisher, count(*) c
FROM pieter.external_occurrence 
GROUP BY publisher
ORDER BY c DESC
```

Let's clear that up a little bit by only looking at the top 20

```
%r
publisher_links_tally <- SparkR::sql("
select publisher, count(*) c
FROM pieter.external_occurrence 
GROUP BY publisher
ORDER BY c DESC")



other_c <-  SparkR::collect(publisher_links_tally) %>%
  filter(c < 10000) %>%
  pull(c) %>%
  sum()

head(publisher_links_tally,20) %>% dplyr::add_row(publisher = "other",c = other_c) %>%
 SparkR::as.DataFrame() %>%
 display
```

Let's have a look at a few friends:

```
%r
SparkR::collect(publisher_links_tally) %>%
 filter(publisher == "European Nucleotide Archive (EMBL-EBI)" | publisher == "ZooKeys" | publisher == "Plazi.org taxonomic treatments database" | publisher == "The International Barcode of Life Consortium") %>%
 SparkR::as.DataFrame() %>%
 display
```

### What taxa are most connected to other collections? 

Let's have a look at what taxa were clustered most with occurrences from other collections.

```
%sql
select taxonkey, count(*) c
FROM pieter.external_occurrence
GROUP BY taxonkey
ORDER BY c DESC
```

```
%r
# Tally the Species that got clustered to other datasets

clustered_species <- SparkR::sql("SELECT taxonkey, count(*) c FROM pieter.external_occurrence GROUP BY taxonkey ORDER BY c DESC")
```

```
%r
# let's look at the most common taxa first
top_500_clustered_species <- head(clustered_species,500)
```

```
%r
# matching with gbif taxonomic backbone to get higher rank information and labels
top_500_clustered_species_parsed <- map_dfr(pull(top_500_clustered_species,taxonkey), function(x) rgbif::name_usage(key = x)$data) %>%
        right_join(top_500_clustered_species, by = c("key" = "taxonkey")) %>% SparkR::as.DataFrame()
```

What taxa are being linked the most?

```
%r
top_500_clustered_species_parsed %>% display()
```

Seems like the Fungi are linking up the most at the moment

```
%r
library(ggplot2)

top_500_clustered_species_parsed %>% 
 SparkR::collect() %>%
 filter(!(issues %in% "")) %>%
 ggplot() +
 aes(x = kingdom, fill = order, weight = c) +
 geom_bar() +
 scale_fill_hue(direction = 1) +
 theme_minimal()
```

### What regions are the occurences we are linking to from?

```
%r
install.packages("countrycode")
```

```
%r
countries <- SparkR::sql("
SELECT publisher, countrycode, count(*) c
FROM pieter.external_occurrence  
GROUP BY countrycode, publisher")

countries %>% 
  SparkR::collect() %>%
  filter(countrycode != 'ZZ') %>%
  mutate(countrycode = countrycode::countrycode(countrycode, origin = "iso2c", destination = "iso3c")) %>%
  SparkR::as.DataFrame() %>%
  display()

```

Let's have a look at the log of the count

```
%r
countries %>% 
  SparkR::collect() %>%
  filter(countrycode != 'ZZ') %>%
  mutate(countrycode = countrycode::countrycode(countrycode, origin = "iso2c", destination = "iso3c")) %>%
  mutate(log_count = log(c)) %>%
  SparkR::as.DataFrame() %>%
  display()
```
